```python
import pandas as pd
tweets_df = pd.read_csv("/content/twitter_training.csv")
tweets_df.head()
```

```python
tweets = tweets_df["im getting on borderlands and i will murder you all ,"].to_list()
Sentiment = []
sentiment_labels = ['Negative', 'Neutral', 'Positive']

for tweet in tweets:
  tokens = tokenizer(str(tweet), return_tensors='pt', padding=True, truncation=True, max_length=512)
  outputs = model(**tokens)
  predictions = outputs.logits
  probabilities = predictions.softmax(dim=1)
  predicted_class = torch.argmax(probabilities, dim=1).item()
  predicted_sentiment = sentiment_labels[predicted_class]
  Sentiment.append(predicted_sentiment)
```

```python
# Assuming you have your sentiment predictions in predictions
sentiment_df = pd.DataFrame({
    'tweet': tweets[:len(Sentiment)],
    'sentiment': Sentiment
})
```
